# Is this entry a valid color or not?


## Summary.
Quick way to validate if a string is a color or not.

Based on [Check if a string defines a color](https://stackoverflow.com/questions/42876366/check-if-a-string-defines-a-color).

**Dependencies:**

`matplotlib`

This is a short notebook that shows a potential application if you want to automate some plots with matplotlib and you want to be sure that a valid color is provided.

```python
from matplotlib.colors import is_color_like

is_color_like("#fcba03")
```

## Potential application.

Validate if the provided `string` inside a visualization is a color or

```python
from typing import Union

def apply_color(color:Union[str,tuple,list]):
  """
  Checks if color  is a valid color or not.

  It raises an exception in case the color is invalid
  """

  if not is_color_like(color):
    raise ValueError(f"Sorry, the value provided as a color, '{color}', is not a valid color.")

  return color
```

Applying our function in one visualization.

# Invalid case.

Inserting an invalid color as string

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')

x = np.linspace(0,20,100)
y = 3*(x+8*np.random.rand(100)) + 12

plt.scatter(x=x,y=y, c=apply_color("No color"))
plt.show()
```

Inserting an invalid tuple as color:

```python
plt.scatter(x=x,y=y, c=apply_color((9999,1,3)))
plt.show()
```

# Valid color.

Inserting a valid string as color:

```python
fig,ax = plt.subplots(1,dpi=96,figsize=(11.25,11.25))
plt.scatter(x=x,y=y, c=apply_color("#fcba03"),s=250,
alpha=0.8)

#Annotations:
plt.xlabel('X axis', size=18)
plt.ylabel('Y axis', size=18, loc="top")
plt.title('Applying a valid color.', size=26,x=0.07,y=1.07)

#Changing position of axes titles.
ax.xaxis.set_label_coords(0,-.10)
ax.yaxis.set_label_coords(-.10,0.97)


plt.show()
```
